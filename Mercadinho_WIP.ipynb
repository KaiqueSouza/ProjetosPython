```python
# -           ____     ____ -
# - |    /   /    \   |     -   
# - |__ /   |      |  |___  -
# - |   \   |      |      | - 
# - |    \   \__\_/   ____| - 
# -              \          -

#WIP = Work in Progress / Trabalho em processo = TEP?

#sistema que simula mercado, controle e manutenção de estoque, produto, quantidade, valor e compra de produto 

#listas de estoque
estoque = ['Caixa café'] 
quantidades = [5] 
valores = [11.59]
#lista de compra
carrinho = [] 
   
#  - - - - - - - - - - - - - - - - - - - - - - - - - - - - funcoes
def adicionar(): #add item no estoque
   yqp = 0
   xqp = int(input("\n Quantidade de produtos que deseja adicionar ao estoque: "))
   print("\n Digite os produtos")  

   while yqp < xqp:
     produto = str(input("\nProduto: "))
     quant = int(input("Quantidade: "))
     preco = float(input("Preco unitario: "))

     estoque.append(produto)
     quantidades.append(quant) #adicionar algum item de acordo com os inputs recebidos 
     valores.append(preco)
     yqp = yqp + 1

   print("\n - Avalie o estoque: - ")
   print("\nProdutos: ", *estoque, sep=' ') 
   print("Quantidades: ", *quantidades, sep=' ')
   print("Valores: ", *valores, sep=' ')

def remover(): #remov item no estoque
   print("\n - Avalie o estoque: - ")
   print("\nProdutos: ", *estoque, sep=', ') 
   print("Quantidades: ", *quantidades, sep=', ')
   print("Valores: ", *valores, sep=', ')

   remov = int(input("\nNumero do item que deseja remover: "))
   estoque.remove(estoque[remov])
   quantidades.remove(quantidades[remov]) #remove item com base no indice dentro da lista
   valores.remove(valores[remov]) 

   print("\nProdutos: ", *estoque, sep=', ') 
   print("Quantidades: ", *quantidades, sep=', ')
   print("Valores: ", *valores, sep=', ')

def comprar(): #compra dos itens
   print("\nProdutos: ", *estoque, sep=', ') 
   print("Quantidades: ", *quantidades, sep=', ')
   print("Valores: ", *valores, sep=', ')

#  - - - - - - - - - - - - - - - - - - - - - - - - - - - - 

while True:
 print("\n | - -- - -- - | SISTEMA MINI MERCADO | - -- - -- - |")

 print("\n- Opições: ") #menu principal 
 menuopicao = str(input("\n(1) Manutenção de estoque \n(2) Compra \n"))


 if menuopicao == '1': #menu secundario
   print("\n- Opições de manutenção de estoque: ")
   menu2opicao = str(input("\n (1) Adicionar item \n (2) Remover item\n"))
   if menu2opicao == '1':
     adicionar()
   elif menu2opicao == '2':
     remover()
   else:
     print("\nOpição invalida")
 elif menuopicao == '22': 
  comprar()
 else:
   print("\nOpição invalida")
```
